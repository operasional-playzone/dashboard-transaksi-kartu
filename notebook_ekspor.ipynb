Terapkan mapping → kolom baru

```python
df['CENTER_STD'] = df['Center'].map(DICT_TOKO_AKTIF)
```

Flag Toko Aktif vs Non Aktif

```python
df['IN_SCOPE_TOKO'] = df['CENTER_STD'].notna()
```

```python
df['IN_SCOPE_TOKO'].value_counts()
```

Audit Cepat Toko

Toko Aktif

```python
df_aktif = (
    df[df['IN_SCOPE_TOKO']]
    [['Center', 'CENTER_STD']]
    .drop_duplicates()
    .sort_values('CENTER_STD')
)

df_aktif
```

Toko Non Aktif

```python
df_non_aktif = (
    df[~df['IN_SCOPE_TOKO']]
    [['Center']]
    .drop_duplicates()
    .sort_values('Center')
)

df_non_aktif
```

Gunakan Nama Center Standar

```python
df.loc[df['IN_SCOPE_TOKO'], 'Center_Final'] = df['CENTER_STD']
df.loc[~df['IN_SCOPE_TOKO'], 'Center_Final'] = df['Center']
```

Buat Kolom Tanggal (Bulanan)

```python
map_bulan = {
    'Januari':1,'Februari':2,'Maret':3,'April':4,
    'Mei':5,'Juni':6,'Juli':7,'Agustus':8,
    'September':9,'Oktober':10,'November':11,'Desember':12
}

df['Bulan'] = df['Bulan'].astype(str).str.title()
df['Bulan_Angka'] = df['Bulan'].map(map_bulan)

df['Tanggal'] = pd.to_datetime(
    dict(year=df['Tahun'], month=df['Bulan_Angka'], day=1),
    errors='coerce'
)
```

Final Dataset untuk Ekspor

```python
df_mesin_aktif = df[df['IN_SCOPE_TOKO']].copy()
df_mesin_non_aktif = df[~df['IN_SCOPE_TOKO']].copy()
```

Ekspor Data Mesin – TOKO AKTIF

```python
df_mesin_aktif.to_excel(
    "DATA_MESIN_TOKO_AKTIF_2024-2025.xlsx",
    index=False
)
```

Ekspor Data Mesin – TOKO NON AKTIF

```python
df_mesin_non_aktif.to_excel(
    "DATA_MESIN_TOKO_NON_AKTIF_2024-2025.xlsx",
    index=False
)
```
